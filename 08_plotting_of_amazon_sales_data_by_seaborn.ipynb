```python
# Remove the currency sign and commas from the 'discounted_price' column and convert it to numeric
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
# Remove the currency sign and commas from the 'actual_price' column and convert it to numeric
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

```python

# Remove any non-numeric characters except for the decimal point
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').str.replace(',', '').str.replace(r'[^\d.]+', '')

# Convert the column to floating-point numbers
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce') / 100

# Print the updated column
print(df['discount_percentage'])
```

```python
import pandas as pd


# Convert the 'rating' column to float, handling non-numeric values
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Print the updated column
print(df['rating'])
```

```python
df.info()
```

```python


# Remove the comma sign from the 'rating_count' column
df['rating_count'] = df['rating_count'].str.replace(',', '')

# Convert the 'rating_count' column to numeric
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Calculate the mean of 'rating_count'
mean_rating_count = df['rating_count'].mean()

# Fill missing values with the mean
df['rating_count'] = df['rating_count'].fillna(mean_rating_count)

# Convert the 'rating_count' column to integer
df['rating_count'] = df['rating_count'].astype(int)

# Print the updated column
print(df['rating_count'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'rating' and 'rating_count' are the names of the columns in your DataFrame 'df'

# Select the columns of interest
columns_of_interest = ['rating', 'rating_count']

# Compute the correlation matrix
corr_matrix = df[columns_of_interest].corr()

# Create the heatmap
sns.heatmap(data=corr_matrix, annot=True, cmap='coolwarm')

# Add a title
plt.title('Correlation Heatmap - Rating vs Rating Count')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'category' and 'rating' are the names of the columns in your DataFrame 'df'

# Get the top 10 categories by their occurrence count
top_10_categories = df['category'].value_counts().nlargest(10).index

# Create a filtered DataFrame with only the top 10 categories
df_top_10_categories = df[df['category'].isin(top_10_categories)]

# Calculate the percentage of each category
category_percentage = (
    df_top_10_categories['category'].value_counts(normalize=True) * 100
).reset_index()

# Rename the columns
category_percentage.columns = ['category', 'percentage']

# Define colors for the bars
colors = sns.color_palette('Set2', len(category_percentage))

# Create the bar plot
plt.figure(figsize=(16, 8))  # Adjust the figure size if needed
ax = sns.barplot(
    data=df_top_10_categories,
    x='category',
    y='rating',
    ci=None,
    palette=colors
)

# Add percentages to the bars
for i, p in enumerate(ax.patches):
    height = p.get_height()
    width = p.get_width()
    x = p.get_x()
    y = height + 0.5  # Adjust the vertical position of the percentage
    ax.text(
        x + width / 2,
        y,
        f"{category_percentage['percentage'][i]:.2f}%",
        ha='center',
        va='center',
        rotation=90,
        color='black'
    )

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('Comparison of Rating across Top 10 Categories')

# Rotate x-axis labels if needed
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()  # Adjust the layout if needed
plt.show()
```
